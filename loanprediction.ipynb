```python
from sklearn.preprocessing import LabelEncoder
encode=LabelEncoder()
```

```python
data.Loan_Status=encode.fit_transform(data.Loan_Status)
```

```python
data.dropna(how='any',inplace=True)
```

```python
from sklearn.model_selection import train_test_split
train ,test=train_test_split(data,test_size=0.2,random_state=0)
```

```python
train_x=train.drop(columns=['Loan_ID','Loan_Status'],axis=1)
train_y=train['Loan_Status']
```

```python
test_x=test.drop(columns=['Loan_ID','Loan_Status'],axis=1)
test_y=test['Loan_Status']
```

```python
train_x=pd.get_dummies(train_x)
test_x=pd.get_dummies(test_x)
                       
```

```python
print('shape of trainnig data:',train_x.shape)
print('shape of test data:',test_x.shape)
```

```python
from sklearn.linear_model import LogisticRegression
mo=LogisticRegression()
```

```python
mo.fit(train_x,train_y)
```

```python
predict=mo.predict(test_x)
```

```python
print('predicted values in test data',predict)
```

```python
print('\n\nAccuracy score on test data:\n\n')
```

```python
 from  sklearn.metrics import accuracy_score
print(accuracy_score(test_y,predict))
```

```python
mo.score(test_x,test_y)
```

```python
mo.fit(train_x,train_y)
```
